```python
import numpy as np
import pandas as pd
import pickle

class PatientQueue:
    def __init__(self, model_file):
        self.queue = []
        self.model = self.load_model(model_file)

    def load_model(self, model_file):
        try:
            with open(model_file, 'rb') as f:
                model = pickle.load(f)
            return model
        except FileNotFoundError:
            print("Error: Model file not found.")
            return None
        except Exception as e:
            print(f"Error loading model: {e}")
            return None

    def insert(self, patient_features):
        if self.model is None:
            print("Error: Model not loaded.")
            return

        try:
            patient_features = np.array(patient_features).reshape(1, -1)
            print("Patient features:", patient_features)
            if hasattr(self.model, 'predict'):
                predicted_class = self.model.predict(patient_features)
                seriousness_level = predicted_class[0]
                print("Seriousness level:", seriousness_level)
                self.queue.append((seriousness_level, patient_features))
                self.queue.sort(key=lambda x: x[0], reverse=True)
            else:
                print("Error: Model does not have 'predict' attribute.")
        except Exception as e:
            print(f"Error inserting patient: {e}")

    def pop(self):
        if not self.queue:
            print("Queue is empty.")
            return None
        print("Before pop, queue contents:", self.queue)
        patient_info= self.queue.pop(0)[1]
        print("After pop, queue contents:", self.queue)
        return patient_info

# Load the trained model
loaded_model = 'retrained_decision_tree_model.pkl'

# Initialize priority queue
user_queue = PatientQueue(loaded_model)

# Continuously accept patient inputs
while True:
    try:
        # Collect patient information
        diastolic_bp = float(input("Enter Diastolic Blood Pressure (mmHg): "))
        systolic_bp = float(input("Enter Systolic Blood Pressure (mmHg): "))
        blood_sugar = float(input("Enter Blood Sugar Level (mg/dL): "))
        oxygen_level = float(input("Enter Oxygen Level (%): "))
        heart_rate = float(input("Enter Heart Rate (beats/min): "))
        body_temp = float(input("Enter Body Temperature (F): "))
        breath_rate = float(input("Enter Breathing Rate (breaths/min): "))
        urinalysis_ph = float(input("Enter Urinalysis (pH): "))
        peak_flow = float(input("Enter Peak Flow: "))
        hydration_level = float(input("Enter Hydration Level (%): "))

        # Create patient feature vector
        patient_features = [diastolic_bp, systolic_bp, blood_sugar, oxygen_level,
                            heart_rate, body_temp, breath_rate, urinalysis_ph,
                            peak_flow, hydration_level]

        # Insert patient into the priority queue
        user_queue.insert(patient_features)

        # Ask if the user wants to insert another patient
        continue_input = input("Do you want to insert another patient? (yes/no): ").lower()
        if continue_input != 'yes':
            break
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        continue

# Process patients from the priority queue
with open("write.txt", "w") as file:
    patient_counter = 1
    while True:
        patient_info = user_queue.pop()
        if patient_info is not None:
            condition = patient_info[0]  # Access the first element which contains the seriousness level
            file.write(f"Condition: {condition} - Patient {patient_counter} - Token No.: {patient_counter}\n")
            patient_counter += 1
        else:
            print("Error: Failed to retrieve patient from queue.")
            break
```
